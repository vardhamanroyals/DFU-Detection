```python

import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)



# 1. IMPORTS
# ==============================
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score

import tensorflow as tf
from tensorflow.keras import layers, models

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# ==============================
# 2. FIXED DATASET PATH (YOUR DATASET)
# ==============================
BASE_PATH = "/kaggle/input/datasets/vuppalaadithyasairam/thermography-images-of-diabetic-foot/ThermoDataBase"

TRAIN_DIR = os.path.join(BASE_PATH, "train")
VAL_DIR   = os.path.join(BASE_PATH, "val")

IMG_SIZE = 128

# ==============================
# 3. SAFE CLASS LOADER (NO ERROR)
# ==============================
def load_paths_labels(base_dir):
    paths, labels = [], []
    
    classes = sorted([d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))])
    print("Detected Classes:", classes)
    
    for label, cls in enumerate(classes):
        folder = os.path.join(base_dir, cls)
        
        for file in os.listdir(folder):
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                paths.append(os.path.join(folder, file))
                labels.append(label)
    
    return np.array(paths), np.array(labels), classes


train_paths, train_labels, class_names = load_paths_labels(TRAIN_DIR)
val_paths, val_labels, _ = load_paths_labels(VAL_DIR)

print("Train:", len(train_paths), "Val:", len(val_paths))


# ==============================
# 4. IMAGE LOADER
# ==============================
def load_images(paths):
    images = []
    for p in tqdm(paths):
        img = cv2.imread(p)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img / 255.0
        images.append(img)
    return np.array(images)

X_train = load_images(train_paths)
X_val   = load_images(val_paths)

y_train = train_labels
y_val   = val_labels


# ==============================
# 5. DIP FEATURE (HOTSPOT RATIO)
# ==============================
def hotspot_ratio(img):
    gray = cv2.cvtColor((img*255).astype(np.uint8), cv2.COLOR_BGR2GRAY)
    
    thresh = np.percentile(gray, 90)
    hotspot = gray > thresh
    
    ratio = np.sum(hotspot) / gray.size
    return ratio

# compute DIP predictions
train_ratios = np.array([hotspot_ratio(img) for img in X_train])
val_ratios   = np.array([hotspot_ratio(img) for img in X_val])

# threshold optimization
best_thresh = 0
best_acc = 0

for t in np.linspace(0.01, 0.5, 50):
    preds = (train_ratios > t).astype(int)
    acc = accuracy_score(y_train, preds)
    
    if acc > best_acc:
        best_acc = acc
        best_thresh = t

print("Best DIP Threshold:", best_thresh)

dip_preds = (val_ratios > best_thresh).astype(int)

print("\nDIP Results:")
print(classification_report(y_val, dip_preds))


# ==============================
# 6. CNN MODEL (RGB)
# ==============================
def build_cnn(input_shape):
    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D(),
        
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(),
        
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.MaxPooling2D(),
        
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    return model

cnn_rgb = build_cnn((IMG_SIZE, IMG_SIZE, 3))

history_rgb = cnn_rgb.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10,
    batch_size=16
)

rgb_preds = (cnn_rgb.predict(X_val) > 0.5).astype(int)

print("\nCNN RGB Results:")
print(classification_report(y_val, rgb_preds))


# ==============================
# 7. HYBRID (5-CHANNEL DIP MAP)
# ==============================
def create_dip_map(img):
    gray = cv2.cvtColor((img*255).astype(np.uint8), cv2.COLOR_BGR2GRAY)
    
    ch1 = gray / 255.0
    ch2 = cv2.GaussianBlur(gray, (5,5), 0) / 255.0
    ch3 = cv2.Canny(gray, 50, 150) / 255.0
    
    thresh = gray > np.percentile(gray, 90)
    ch4 = thresh.astype(float)
    
    ch5 = cv2.Laplacian(gray, cv2.CV_64F)
    ch5 = np.abs(ch5) / np.max(np.abs(ch5))
    
    stacked = np.stack([ch1, ch2, ch3, ch4, ch5], axis=-1)
    return stacked

X_train_h = np.array([create_dip_map(img) for img in X_train])
X_val_h   = np.array([create_dip_map(img) for img in X_val])

cnn_hybrid = build_cnn((IMG_SIZE, IMG_SIZE, 5))

history_h = cnn_hybrid.fit(
    X_train_h, y_train,
    validation_data=(X_val_h, y_val),
    epochs=10,
    batch_size=16
)

hybrid_preds = (cnn_hybrid.predict(X_val_h) > 0.5).astype(int)

print("\nHYBRID Results:")
print(classification_report(y_val, hybrid_preds))


# ==============================
# 8. CONFUSION MATRIX
# ==============================
def plot_cm(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    
    plt.imshow(cm, cmap='Blues')
    plt.title(title)
    plt.colorbar()
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

plot_cm(y_val, dip_preds, "DIP")
plot_cm(y_val, rgb_preds, "CNN RGB")
plot_cm(y_val, hybrid_preds, "HYBRID")


# ==============================
# 9. TRAINING GRAPHS
# ==============================
def plot_history(history, title):
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(title + " Accuracy")
    plt.legend(['Train','Val'])
    plt.show()

plot_history(history_rgb, "CNN RGB")
plot_history(history_h, "HYBRID")


# ==============================
# 10. VISUAL COMPARISON
# ==============================
plt.figure(figsize=(12,6))

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(X_val[i])
    plt.title(f"True:{y_val[i]} DIP:{dip_preds[i]} CNN:{rgb_preds[i]} HYB:{hybrid_preds[i]}")
    plt.axis('off')

plt.suptitle("Model Comparison")
plt.show()

# Use the kagglehub client library to attach Kaggle resources like competitions, datasets, and models to your session
# Learn more about kagglehub: https://github.com/Kaggle/kagglehub/blob/main/README.md

import kagglehub
# kagglehub.dataset_download('<owner>/<dataset-slug>')
```

```python
# ==============================
# 1. IMPORTS
# ==============================
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import tensorflow as tf
from tensorflow.keras import layers, models

# ==============================
# 2. DATA PATH
# ==============================
BASE_PATH = "/kaggle/input/datasets/vuppalaadithyasairam/thermography-images-of-diabetic-foot/ThermoDataBase"

TRAIN_DIR = os.path.join(BASE_PATH, "train")
VAL_DIR   = os.path.join(BASE_PATH, "val")

IMG_SIZE = 128

# ==============================
# 3. LOAD DATA PATHS
# ==============================
def load_paths_labels(base_dir):
    paths, labels = [], []
    
    classes = sorted([d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))])
    print("Classes:", classes)
    
    for label, cls in enumerate(classes):
        folder = os.path.join(base_dir, cls)
        
        for file in os.listdir(folder):
            if file.lower().endswith(('.png','.jpg','.jpeg')):
                paths.append(os.path.join(folder, file))
                labels.append(label)
                
    return np.array(paths), np.array(labels), classes

train_paths, train_labels, class_names = load_paths_labels(TRAIN_DIR)
val_paths, val_labels, _ = load_paths_labels(VAL_DIR)

print("Train:", len(train_paths), "Val:", len(val_paths))


# ==============================
# 4. LOAD IMAGES
# ==============================
def load_images(paths):
    images = []
    for p in tqdm(paths):
        img = cv2.imread(p)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img / 255.0
        images.append(img)
    return np.array(images)

X_train = load_images(train_paths)
X_val   = load_images(val_paths)

y_train = train_labels
y_val   = val_labels


# ==============================
# 5. DIP (RED + YELLOW HOTSPOT)
# ==============================
def extract_hotspot(img):
    img_uint8 = (img * 255).astype(np.uint8)
    hsv = cv2.cvtColor(img_uint8, cv2.COLOR_BGR2HSV)
    
    # RED
    r1 = cv2.inRange(hsv, (0,120,120), (10,255,255))
    r2 = cv2.inRange(hsv, (170,120,120), (180,255,255))
    red = cv2.bitwise_or(r1, r2)
    
    # YELLOW
    yellow = cv2.inRange(hsv, (20,120,120), (35,255,255))
    
    mask = cv2.bitwise_or(red, yellow)
    
    # clean noise
    kernel = np.ones((3,3), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    
    binary = (mask > 0).astype(np.uint8)
    ratio = np.sum(binary) / binary.size
    
    return ratio, mask

# compute ratios
train_ratios = np.array([extract_hotspot(img)[0] for img in X_train])
val_ratios   = np.array([extract_hotspot(img)[0] for img in X_val])

# threshold tuning
best_thresh, best_acc = 0, 0
for t in np.linspace(0.01, 0.5, 50):
    preds = (train_ratios > t).astype(int)
    acc = accuracy_score(y_train, preds)
    if acc > best_acc:
        best_acc = acc
        best_thresh = t

dip_preds = (val_ratios > best_thresh).astype(int)

print("\nDIP Results:")
print(classification_report(y_val, dip_preds))


# ==============================
# 6. CNN MODEL
# ==============================
def build_model(input_shape):
    model = models.Sequential([
        layers.Conv2D(32,(3,3),activation='relu',input_shape=input_shape),
        layers.MaxPooling2D(),
        layers.Conv2D(64,(3,3),activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(128,(3,3),activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128,activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1,activation='sigmoid')
    ])
    
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

cnn = build_model((IMG_SIZE, IMG_SIZE, 3))

history = cnn.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10,
    batch_size=16
)

cnn_preds = (cnn.predict(X_val) > 0.5).astype(int)

print("\nCNN Results:")
print(classification_report(y_val, cnn_preds))


# ==============================
# 7. HYBRID MODEL
# ==============================
def create_features(img):
    gray = cv2.cvtColor((img*255).astype(np.uint8), cv2.COLOR_BGR2GRAY)
    
    ch1 = gray/255.0
    ch2 = cv2.GaussianBlur(gray,(5,5),0)/255.0
    ch3 = cv2.Canny(gray,50,150)/255.0
    ch4 = (gray > np.percentile(gray,90)).astype(float)
    
    lap = cv2.Laplacian(gray,cv2.CV_64F)
    ch5 = np.abs(lap)/(np.max(np.abs(lap))+1e-6)
    
    return np.stack([ch1,ch2,ch3,ch4,ch5],axis=-1)

X_train_h = np.array([create_features(img) for img in X_train])
X_val_h   = np.array([create_features(img) for img in X_val])

cnn_h = build_model((IMG_SIZE, IMG_SIZE, 5))

history_h = cnn_h.fit(
    X_train_h, y_train,
    validation_data=(X_val_h, y_val),
    epochs=10,
    batch_size=16
)

hyb_preds = (cnn_h.predict(X_val_h) > 0.5).astype(int)

print("\nHYBRID Results:")
print(classification_report(y_val, hyb_preds))


# ==============================
# 8. CONFUSION MATRIX
# ==============================
def plot_cm(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.imshow(cm, cmap='Blues')
    plt.title(title)
    plt.colorbar()
    plt.show()

plot_cm(y_val, dip_preds, "DIP")
plot_cm(y_val, cnn_preds, "CNN")
plot_cm(y_val, hyb_preds, "HYBRID")


# ==============================
# 9. TRAINING GRAPH
# ==============================
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("CNN Accuracy")
plt.legend(['Train','Val'])
plt.show()


# ==============================
# 10. HOTSPOT VISUALIZATION
# ==============================
plt.figure(figsize=(12,6))

for i in range(6):
    img = X_val[i]
    _, mask = extract_hotspot(img)
    
    overlay = img.copy()
    overlay[mask > 0] = [1,0,0]
    
    plt.subplot(2,3,i+1)
    plt.imshow(overlay)
    plt.title(f"T:{y_val[i]} D:{dip_preds[i]} C:{cnn_preds[i]} H:{hyb_preds[i]}")
    plt.axis('off')

plt.suptitle("Hotspot + Model Comparison")
plt.show()
```
